# Method and Plan

Regression using the k-nearest neighbours algorithm will be used to predict the quantitative variable,  “played_hours,” using the provided demographics (3 predictors: experience, gender, and age). From the preliminary data analysis and visualization, not all variables have a linear relationship with “played_hours,” therefore, linear regression would have reduced accuracy vs. k-nearest neighbours. KNN performance decreases when the data is sparse, scarce, and contains outliers. I will assume the dataset is large enough for KNN to be “good enough.” A limitation I will look out for is the fact that KNN does not work with categorical data as predictors (“experience” and “gender” variables), so I will try to wrangle the data into something that works. If KNN seems impossible, I may switch methods.

-I will create the recipe and model using the training data, then combine them into a workflow. \
-The model will be tuned with cross-validation and tune_grid() to select the optimal k-value with minimal RMSE (Root Mean Square Error). \
